```python
mg = 4
days = 7
Ad1 = np.ones((mg,mg))
Ad2 = np.eye(mg)

# print(Ad1)
# print(Ad2)
```

```python
Ad1 = np.kron(np.eye(days), np.ones((mg,mg)))
# print(Ad1)
Ad00 = np.zeros((np.shape(Ad1)[1],mg))
Ad01 = np.zeros((mg,np.shape(Ad1)[0]))
# print(Ad00)
# print(Ad01)
# print(np.shape(Ad1)[1])
# print(np.shape(Ad1)[0])
Ad = np.block([
    [Ad00,            Ad1],
    [np.zeros((mg,mg)), Ad01]
])
print(Ad)
```

```python
G0=nx.from_numpy_matrix(Ad)
plt.plot
# nx.draw(G0)
nx.draw_spring(G0)
plt.show()
```
